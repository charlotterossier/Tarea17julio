```python
df=pandas.read_html(t, header=0)[0]
df.head()
```

```python
d=df.to_dict(orient="split")
d
```

```python
df=df.dropna(how="all").reset_index()
df
```

```python
df.keys()
```

```python
datosTablaFiltrada=df.filter(items=['index', 'Pos \tPosición', 'Selección', 'Mundiales'])
datosTablaFiltrada
```

```python
datos=datosTablaFiltrada.to_dict(orient='split')
```

```python
datosTabla=datos['data']
datosTabla=datosTabla[:15]
datosTabla
```

```python
l=[]
for index, posicion, seleccion, mundiales in datosTabla:
    lugar= seleccion + "," + mundiales
    print(lugar)
    latlng=geocoder.google(lugar).latlng
    print(latlng)
    seleccion =[posicion, seleccion, mundiales, latlng]
    l.append(seleccion)
```

```python
l[0]
```

```python
len(l)
```

```python
l=[[posicion, seleccion, mundiales, ubicacion]
    for posicion, seleccion, mundiales, ubicacion in l if ubicacion !=None]
```

```python
len(l)
```

```python
m=folium.Map(
    location=[-15.7929155, -47.8940685],
    tiles="Stamen Toner",
    zoom_start=9
)
for posicion, seleccion, mundiales, ubicacion in l:
    folium.Marker(ubicacion, popup=seleccion).add_to(m) 
```

```python
m=folium.Map(
    location=[-15.7929155, -47.8940685],
    tiles="Stamen Toner",
    zoom_start=9
)
for posicion, seleccion, mundiales, ubicacion in l:
    folium.Circle(ubicacion, 
                  popup=seleccion,
                 color="lightblue",
                 fill=True,
                 radius=(mundiales)*3
                 ).add_to(m)
m
```
